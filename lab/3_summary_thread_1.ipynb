```python
import re

def extract_latest_email_content(email_text):
    email_text = str(email_text)
    # Regular expression pattern to match the start of a previous email in the thread
    pattern = re.compile(
        r"(^On .{1,1000}?\n.{0,1000}?>|^Em .{1,1000}?\n.{0,1000}?>|^De .{1,1000}?\n.{0,1000}?>|^From .{1,1000}?\n.{0,1000}?>)",
        re.MULTILINE | re.DOTALL
    )
    
    # Split the email text based on the pattern
    parts = re.split(pattern, email_text)
    
    # The latest email content is the first part before any split
    latest_email_content = parts[0].strip()
    
    return latest_email_content

# Example emails
email1 = """
Eu troquei o TNSnames na última vez que deu problema na Agenda SAT.
Descobri que não estava substituindo pela GPO e fiz manualmente.
Como o email do Felipe foi enviado após o fim do meu expediente, só vi esse
email hoje.

Em seg., 15 de jan. de 2024 às 10:18, Paulo Ricardo Ritter <
pritter@trt4.jus.br> escreveu:

>

A GPO de atualização do TNSnames (desktop-all-oracle) nunca rodou no
MarkVision, porque a estação não  está no escopo da aplicação da GPO

*trt4.gov.br/infraestrutura/DIM/MARKVISION
*
"""

email2 = """
Yes that is fine, I will email you in the morning.

On Fri, Nov 16, 2012 at 1:48 PM, Zapier <contact@zapier.com> wrote:

> Our support team just commented on your open Ticket:
> "Hi Royce, can we chat in the morning about your question?"
"""

# Extract the latest email content
print(extract_latest_email_content(email1))
print("------")
print(extract_latest_email_content(email2))
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo-0125")
```

```python
from langchain.prompts import PromptTemplate

# The initial summary should be under {max_words} words and contain {entity_range} informative Descriptive Entities from the {content_category}.
    # - Your summary should consist of a list of senteces. Each sentence should summarize an information present in the email.
    # - Remember, to keep each summary to max {max_words} words.

prompt = PromptTemplate.from_template(
    """
    As an expert copy-writer, you will write a summary of the user provided email. 
    
    # Guidelines
    - First read through all the below sections to get an understanding of the task.
    - The emails you will summarize comprise of conversations of IT Technicians talking about a problem, asking for help, explaninig how to use a technology, etc.
    - The conversation could happen in a lot of emails and in this case, you will summarize just one email of the conversation.
    - Your summary will be used to create a knowledge base document of a IT company to be used by IT Technicians. 
    - Your summary will be used in conjunction with others summaries of others emails of this conversations, 
            so each summary should have all the important information that could be important to understand later emails.
    - you should remove all the words that not contain any important information to create a knowledge base document or a manual.
    - At the same time, you should maintain all the words and information that could be necessary to work with the technologies described in the email and to understande later emails.
    - Important information to create a knowledge base document: 
            - a summary of the problem being addressed, 
            - a exact description char by char of the error message that happened. you should just copy system messages necessary to address the problem.
            - the name of the technology that gave the error message
            - a possible solution, 
            - a technique that didn't work and why it didn't work, 
    - Make every word count.
    - Make space with fusion, compression, and removal of uninformative phrases like "the email discusses".
    - The summaries should be highly dense and concise yet self-contained, e.g., easily understood without the email.
    - The language of the conversation is generaly in Brazilian Portuguese. So your summary should be in Brazilian Portuguese.
    
    - Do not discuss the email itself, focus on the content: informative descriptive entities, and details.

    - Some emails dont have any important information or it are just talking about burocratic things like redirecting email to the correct sector, someone acknoledging something, someone thanking, etc. 
                In the case of the email dont have any important information or it just have burocratic things, you should output: [Não contém informações importantes]


    # Example email:
        '
        Prezada DGBD, boa tarde!

        Estamos com erro ao fazer operações na base do Infor de São Leopoldo. Podem
        dar uma olhada, por favor?


        [2024-01-12 11:41:18] Falha ao conectar … base do inFOR de SÆo Leopoldo 1¦
        > VT (SAOLEO.WORLD) com o usu rio inforadm: Exception calling "Open" with "0"
        > argument(s): "ORA-12541: TNS: nÆo h  listener"




        Atenciosamente,

        Felipe Giacomel
        Divisão de Microinformática
        TRT4 - Secretaria de Tecnologia da Informação e Comunicações
        '

    # Example output summary
        '
        Problema: erro ao utilizar base do Infor de São Leopoldo
        Tecnologia: Oracle
        Mensagem de erro:
        [2024-01-12 11:41:18] Falha ao conectar … base do inFOR de SÆo Leopoldo 1¦
        > VT (SAOLEO.WORLD) com o usu rio inforadm: Exception calling "Open" with "0"
        > argument(s): "ORA-12541: TNS: nÆo h  listener"
        '



    # Output
    - output just raw text of the summarization.

    # EMAIL TO BE SUMMARIZED
    {email_content}
    """
)
```

```python
chain = prompt | llm
```
